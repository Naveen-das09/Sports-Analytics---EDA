```python
df1['Height'].isnull().sum()
```

```python
def Height_in_feet(i):
    split_string=i.split("'")
    float_value=float(split_string[0])+(float(split_string[1]))/12
    return float_value
```

```python
df1['Height']=df1['Height'].apply(Height_in_feet)
```
